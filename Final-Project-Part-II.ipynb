### 3- Construct a query to retrieve the count of case incidents, count and percent (rounded to 3 decimal places) of arrests for each year;  order descending by arrested percent.
<span style="font-size:7px"><b>Hint:</b> 22 rows affected -- data[2015=26.451%] -- parse dates, uses "case when"</span>

```python
%%sql
select year,case_incidents,total,round((100*cast(total as float)/case_incidents)::numeric,3) as percentage
from (select extract(year from incident_date) as year,count(case_number) as case_incidents,
     count(case when arrest=1 then 1 else NULL end) as total
     from cc_cases group by year) as new;
```

### 5- For each year, which month is ranked #1 as having the greatest number of crime incidents with a primary crime description containing the term "NARCOTICS" or secondary crime desciption containing the phrase "GUN"? Display the year, month of the cases incident date, the primary and secondary crime description, the count of incidents and the ranking number. Order by year descending.

<span style="font-size:7px"><b>Hint:</b>22 rows affected -- data[2012-2, incidents=1942] ranking# should all be 1 -- involves parsing dates, a nested query with a window function and groups</span>

```python
%%sql
SELECT * FROM
(SELECT rank() OVER (PARTITION BY year order by Crimes_count DESC), year, month, iucr_primary_desc, iucr_secondary_desc,Crimes_count 
    FROM (SELECT CAST(EXTRACT(YEAR FROM incident_date)AS INT) AS year,DATE_PART('month', incident_date) AS month,  
         iucr_primary_desc, iucr_secondary_desc, COUNT(case_number) AS Crimes_count
         FROM cc_cases 
         JOIN cc_iucr_codes_primary_descriptions USING(iucr_code)
         JOIN cc_iucr_codes_secondary_descriptions USING(iucr_code)
         WHERE iucr_primary_desc LIKE '%NARCOTICS%'
         OR iucr_secondary_desc LIKE '%GUN%'
         GROUP BY iucr_primary_desc, iucr_secondary_desc,year, month
         ORDER BY year) new)ans WHERE rank=1;
```

### 6 What is the average difference (in days, expressed as an integer) between the updated and incident dates for case arrests and primary and secondary crime descriptions ordered by the average difference in days.

<span style="font-size:7px"><b>Hint:</b>HINT: 538 rows affected -- output first row data[NARCOTICS,SALE/DEL DRUG PARAPHERNALIA,False,6] -- Involves Aggregate groups, parse date, calculate date difference</span>

```python
%%sql
SELECT iucr_primary_desc, iucr_secondary_desc, arrest, CAST(AVG(DATE_PART('day', updated_on::timestamp - incident_date::timestamp)) as INT) AS avg_duration
FROM cc_cases
JOIN cc_iucr_codes_primary_descriptions USING(iucr_code)
JOIN cc_iucr_codes_secondary_descriptions USING(iucr_code)
GROUP BY iucr_primary_desc, iucr_secondary_desc, arrest
ORDER BY avg_duration;
```

```python
%%sql
select location_description,count(cc_cases.case_number) as crimes_count
from cc_cases
join cc_case_location using(case_number)
group by location_description
order by crimes_count desc limit 10;
```

```python
%%sql
select iucr_primary_desc, iucr_secondary_desc,location_description
from cc_iucr_codes_primary_descriptions
join cc_cases on cc_cases.iucr_code=cc_iucr_codes_primary_descriptions.iucr_code
join cc_iucr_codes_secondary_descriptions on cc_iucr_codes_secondary_descriptions.iucr_code=cc_cases.iucr_code
join cc_case_location on cc_case_location.case_number=cc_cases.case_number
where cc_cases.case_number in
(select case_number from cc_case_location
 where location_description LIKE '%SCHOOL%')
group by iucr_primary_desc,iucr_secondary_desc,location_description
```

#### Above query combines four tables to get the primary and secondary descriptions of the crimes that happen in locations that involved a school and are arranged by the primary description of the type of crime. This is to see what kind of crimes have happened in public/private school building/ground.

```python
%%sql
select iucr_primary_desc,sum(ans.count) as crime_count from 
(select iucr_primary_desc,count(cc_cases.case_number),location_description
from cc_iucr_codes_primary_descriptions
join cc_cases on cc_cases.iucr_code=cc_iucr_codes_primary_descriptions.iucr_code
join cc_case_location on cc_case_location.case_number=cc_cases.case_number
where cc_cases.case_number in
(select case_number from cc_case_location
 where location_description LIKE '%SCHOOL%')
group by iucr_primary_desc,location_description) as ans
group by iucr_primary_desc
order by crime_count desc
```

#### Above query shows the most frequently reported crimes that happened in locations with schools. This is helpful to create citizen awareness and bring focus on making schools a safe place for everyone. Criminal damage, narcotics, criminal trespass and weapons violation are among the top 4 frequent crimes. Regular checks for vandalism and narcotics can be conducted in school premises to identify and prevent such crimes from happening frequently.
